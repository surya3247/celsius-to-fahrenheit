```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
user_details = pd.read_excel("Data Analyst Intern Assignment - Excel.xlsx", sheet_name="UserDetails.csv")
cooking_sessions = pd.read_excel("Data Analyst Intern Assignment - Excel.xlsx", sheet_name="CookingSessions.csv")
order_details = pd.read_excel("Data Analyst Intern Assignment - Excel.xlsx", sheet_name="OrderDetails.csv")
```

```python
# Data Cleaning
# Check for missing values
print("Missing values in UserDetails:\n", user_details.isnull().sum())
print("Missing values in CookingSessions:\n", cooking_sessions.isnull().sum())
print("Missing values in OrderDetails:\n", order_details.isnull().sum())

# Fill or drop missing values (example: dropping rows with null values)
user_details.dropna(inplace=True)
cooking_sessions.dropna(inplace=True)
order_details.dropna(inplace=True)

```

```python
# Merge datasets
merged_data = pd.merge(user_details, cooking_sessions, on="User ID", how="inner")
merged_data = pd.merge(merged_data, order_details, on="User ID", how="inner")
```

```python
print(merged_data.columns)
```

```python
# Analysis
# Popular dishes
popular_dishes = merged_data['Dish Name_x'].value_counts().head(10)
print("Most Popular Dishes:\n", popular_dishes)

# Relationship between cooking sessions and orders
cooking_orders_relationship = merged_data.groupby('User ID').agg({
    'Dish Name_x': 'count',
    'Dish Name_y': 'count'
}).rename(columns={'Dish Name_x': 'CookingSessions', 'Dish Name_y': 'Orders'})

# Demographic factors
demographics_analysis = merged_data.groupby('Location')['Dish Name_x'].count()
```

```python
# Visualizations
# 1. Popular Dishes
plt.figure(figsize=(10, 6))
popular_dishes.plot(kind='bar', color='skyblue')
plt.title("Top 10 Popular Dishes")
plt.xlabel("Dish Name")
plt.ylabel("Order Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Relationship between Cooking Sessions and Orders
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cooking_orders_relationship, x='CookingSessions', y='Orders')
plt.title("Relationship Between Cooking Sessions and Orders")
plt.xlabel("Cooking Sessions")
plt.ylabel("Orders")
plt.show()
```

```python
# 3. Demographic Analysis
plt.figure(figsize=(8, 6))
demographics_analysis.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Orders by Location")
plt.ylabel("")
plt.show()
```

```python
# Count of dishes prepared
popular_dishes = merged_data['Dish Name_x'].value_counts().head(10)

# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_dishes.values, y=popular_dishes.index, palette="viridis")
plt.title("Top 10 Most Prepared Dishes", fontsize=16)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Dish Name", fontsize=12)
plt.show()
```

```python
# Bar Chart (Vertical)
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_dishes.index, y=popular_dishes.values, palette="viridis")
plt.title("Most Favorite Meals", fontsize=16)
plt.xlabel("Meal Time", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate the dish names for better readability
plt.show()

```
